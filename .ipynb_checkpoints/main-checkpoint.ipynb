```python
import pandas as pd
```

```python
csv = pd.read_csv("contracts26.csv", low_memory = False)
```

```python
states = ['GA', 'LA', 'AL', 'TN', 'NC']
```

```python
len([x for x in csv['recipient_state_code'] if x in states])
```

```python
filtered = csv[csv['recipient_state_code'].isin(states)]
```

```python
avg_by_state = filtered.groupby('recipient_state_code')['federal_action_obligation'].agg(['sum','mean','count'])
avg_by_state
```
